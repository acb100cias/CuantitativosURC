```python
from matplotlib.pyplot import *
```

```python
plot(Z1,f(Z1),label=r'$f(x)=x^2+x-1$',color='red',marker='^',alpha=0.4)
plot(Z1,Z1,color='green',alpha=0.2)
plot(Z1,Z1**2,color='purple',label=r'$g(x)=x^2$')
legend()
```

```python
T=linspace(-10,10,10000)
```

```python
T
```

```python
plot(T,f(T),color='orange')
grid()
```

```python
plot(T,T**3,color='magenta')
grid()
title('Funcion cúbica')
xlabel('x')
ylabel('$x^3$')
```

```python
plot(T,T**9,color='magenta')
grid()
title('Funcion potencia 9')
xlabel('x')
ylabel('$x^9$')
```

```python
def A(x):
  return 4*x+5
plot(T,A(T),color='k')
grid()
title('Funcion afín')
xlabel('x')
ylabel('$4x+5$')
```

```python
plot(T,exp(T),color='k')
grid()
title('Funcion exponencial')
xlabel('x')
ylabel('$e^x$')
```

```python
T1=linspace(-20,20,100000)
plot(T1,cos(T1),color='g',label='cos(t)')
plot(T1,sin(T1),color='r',label='sin(t)')
grid()
legend()
title('Funciones trigonométricas')
savefig('figura1.png')
```
