#Función logistica#

$$
P(0)=P_0\\
P(t)=P_0+f^t(P_0)\\
f(x)=rx(1-\frac{x}{K})
$$

```python
def f(r,K,x):
  return r*x*(1-x/K)
```

```python
r=1.5
K=10
X=np.arange(0,K,0.01)
plt.plot(X,f(r,K,X))
```

```python
PL=[10.1]
for i in range(30):
  PL.append(PL[-1]+f(r,K,PL[-1]))
```

```python
r=4
plt.plot(X,f(r,K,X),label='r=%1f'%r)
plt.plot(X,X,label='y=x')
xc=0.05
for i in range(300):
    plt.plot([xc,xc],[xc,f(r,K,xc)], color='k', linestyle='--')
    plt.plot([f(r,K,xc),xc],[f(r,K,xc),f(r,K,xc)], color='k', linestyle='--')
    xc=f(r,K,xc)
plt.legend()
```

```python
n = 10000
r = np.linspace(1, 3.925, n)
iterations = 1000
last = 100
x = 1e-5 * np.ones(n)
lyapunov = np.zeros(n)
fig,ax1  = plt.subplots(1, 1, figsize=(20, 10))
for i in range(iterations):
    x = f(r,K, x)
    lyapunov += np.log(abs(r - 2 * r * x))
    if i >= (iterations - last):
        ax1.plot(r, x, ',k', alpha=.25)
ax1.set_xlim(1., 3.925)
ax1.set_title('Diagrama de bifurcaciones')
plt.xlabel('r')
plt.ylabel('X')
```
